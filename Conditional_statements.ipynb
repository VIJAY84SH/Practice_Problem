```python
# if - elif - elif - else(elif ladder)

score = 45

if score>=90:
  print('T1')
elif score >= 80:
  print('T2')
elif score >=65:
  print('T3')
else:
  print('T4')
```

```python
grades = int(input('Enter you grades= '))

if grades >= 90:
  print('A')
elif grades >= 80:
  print('B')
elif grades >=70:
  print('C')
elif grades >= 65:
  print('D')
else:
  print('You have failed the exam')
```

```python
# nested if - else:

a = 18
if a >= 20:
  print('Condition is True')
else:
  if a >= 15:
    print('2nd condition is true')
  else:
    print('All conditions are False')
```
